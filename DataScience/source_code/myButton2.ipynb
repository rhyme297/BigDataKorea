```python
from tkinter import *
```

```python
def call(event): #event 자리에는 아무것이나 넣어도 됨
    print("버튼을 누르셨군요.")
    a = 100
    b = 200
    print(a+b)
    l["text"] = "나로 바꿔요."


w = Tk()
w.configure(bg="white")
b = Button(w, text="합계를 구해주세요.")
b.pack()

b.bind("<Button-1>", call) #버튼을 한 번 누르면 앞쪽의 call 함수를 호출한다는 의미

l = Label(w, text="반갑습니다...", bg="white", font=("궁서체", 30))
l.pack



w.mainloop()
```
