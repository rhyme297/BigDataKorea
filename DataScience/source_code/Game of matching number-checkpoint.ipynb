```python
import random

number = random.randint(1,100)
number

print("1부터 100 사이의 숫자를 맞추어주세요.")
tries = 0

while tries < 10:
    guess = int(input("숫자를 입력하세요..."))
    tries = tries + 1
    if guess == number:
        print("정답입니다.")
        break
    else:
        print("정답이 아닙니다.")
        if guess < number:
            print("너무 작음!!!")
        elif guess > number:
            print("너무 큼!!!")
        else:
            print("...")
    
```
